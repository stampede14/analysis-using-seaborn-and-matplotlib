```python
import seaborn as sns
import pandas as pd

# I am just using different seaborn and matplotlib  plots to visualize the data, I am not doing any data cleaning
```

```python
df = pd.read_csv("/kaggle/input/tipping/tips.csv")
```

```python
df.head()
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
#here we can see most of tips are in the range of 2$
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
#using regression to check tip vs totalbill
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
#here we can see on the basis of sex who tipped more
```

```python
sns.distplot(df['tip'])
#dist plot to see the range of tip
```

```python
sns.distplot(df['tip'], kde = False , bins = 10)
```

```python
sns.barplot(y= 'total_bill', x =  'smoker', data =df)
#total_bill by smoker & non smoker
```

```python
sns.boxplot(x = 'day' ,y = 'total_bill',hue = 'sex', data = df)
```

```python
sns.violinplot(x = 'day' ,y = 'total_bill',  data = df)
```
